top 10 state that order the most


```python
Top_states = df.groupby(["State"]).sum().sort_values("Orders", ascending=False).head(10) 
Top_states = Top_states[["Orders"]].round(2)
Top_states.reset_index(inplace=True) 
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(19,7))
plt.title("order on the basis of Zone",fontsize=18)
plt.bar(Top_states["State"],Top_states["Orders"],color='#7FFFD4',edgecolor='Red',linewidth=1)
plt.xlabel("State",fontsize=19)
plt.ylabel("Order",fontsize=19)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)
plt.show

```

```python
Top_Product_Category=df.groupby(["Zone"]).sum().sort_values("Amount",ascending=False).head(8)
Top_Product_Category=Top_Product_Category[["Orders"]].round(2)
Top_Product_Category.reset_index(inplace=True)
total_revenue_products=Top_Product_Category["Orders"].sum()
total_revenue_products = str(int(total_revenue_products))
total_revenue_products = '$' + total_revenue_products
```

```python
plt.rcParams["figure.figsize"]=(13,7)
plt.rcParams['font.size']= 12.0
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#55B4B0','#E15D44','#009B77','#B565A7']
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
fig1, ax1 = plt.subplots()
ax1.pie(Top_Product_Category['Orders'],colors=colors,labels=Top_Product_Category['Zone']Top_Product_Category["Orders"], startangle=90,explode=explode)
centre_circle = plt.Circle((0,0),0.80,fc='white') # Draw a circle on the pie chart
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
label = ax1.annotate('Total Revenue \n of these products \n'+str(total_revenue_products),color = 'red', xy=(0, 0), fontsize=12, ha="center")
plt.tight_layout()
plt.show()
```
