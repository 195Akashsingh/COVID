```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")  # Example dataset



plt.figure(figsize=(8, 4))

# Define a custom color palette (you can use any valid Matplotlib or Seaborn palette)
custom_palette = ["#FF5733", "#33FF57", "#3366FF", "#FF33D1"]

sns.barplot(data=data, x="day", y="total_bill", palette=custom_palette)




```
